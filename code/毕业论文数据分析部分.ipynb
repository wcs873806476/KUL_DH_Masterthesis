```python
import pandas as pd
import pycountry
import pycountry_convert as pc

# Âä†ËΩΩ‰Ω†‰∏ä‰º†ÁöÑÊñá‰ª∂
df = pd.read_excel("‰ΩúËÄÖÂõΩÂÆ∂ÂàÜÂ∏É.xlsx")

# ÂõΩÂÆ∂ÂêçÁß∞Á∫†ÈîôË°®
country_corrections = {
    "UK": "United Kingdom",
    "USA": "United States",
    "Russia": "Russian Federation",
    "South Korea": "Korea, Republic of",
    "North Korea": "Korea, Democratic People's Republic of",
    "Iran": "Iran, Islamic Republic of",
    "Venezuela": "Venezuela, Bolivarian Republic of",
    "Syria": "Syrian Arab Republic",
    "Vietnam": "Viet Nam",
    "Bolivia": "Bolivia, Plurinational State of",
    "Tanzania": "Tanzania, United Republic of"
}

def get_continent(country_name):
    try:
        name = country_corrections.get(country_name, country_name)
        country = pycountry.countries.get(name=name)
        if not country:
            country = pycountry.countries.search_fuzzy(name)[0]
        code = country.alpha_2
        continent_code = pc.country_alpha2_to_continent_code(code)
        return pc.convert_continent_code_to_continent_name(continent_code)
    except:
        return "Unknown"

# Ê∑ªÂä†Â§ßÊ¥≤Âàó
df["Continent"] = df["Country"].apply(get_continent)

# ‰øùÂ≠òÁªìÊûú
df.to_excel("ÂõΩÂÆ∂_Â§ßÊ¥≤_ÁªüËÆ°ÁªìÊûú.xlsx", index=False)
```

```python
import pandas as pd
import networkx as nx
from itertools import combinations
import matplotlib.pyplot as plt

# ËØªÂèñÊï∞ÊçÆ
df = pd.read_csv('2authors.csv', sep=';')  # Ê†πÊçÆ‰Ω†ÁöÑÊñá‰ª∂Ê†ºÂºèË∞ÉÊï¥ÂàÜÈöîÁ¨¶

# ÂéªÈô§Áº∫Â§±È°π
df = df.dropna(subset=['‰ΩúËÄÖ', 'ÈìæÊé•'])

# ÂàùÂßãÂåñÂõæ
G = nx.Graph()

# ÊåâÊñáÁ´†ÈìæÊé•ÂàÜÁªÑÔºåÊûÑÂª∫Âêà‰ΩúËæπ
for link, group in df.groupby('ÈìæÊé•'):
    authors = group['‰ΩúËÄÖ'].unique()
    if len(authors) >= 2:
        for pair in combinations(authors, 2):
            a1, a2 = sorted(pair)
            if G.has_edge(a1, a2):
                G[a1][a2]['weight'] += 1
            else:
                G.add_edge(a1, a2, weight=1)

# ÊâìÂç∞Âêà‰ΩúÁΩëÁªú‰ø°ÊÅØ
print(f"‰ΩúËÄÖËäÇÁÇπÊï∞ÈáèÔºö{G.number_of_nodes()}")
print(f"Âêà‰ΩúËæπÊï∞ÈáèÔºö{G.number_of_edges()}")

# ÊòæÁ§∫Âêà‰ΩúÊúÄÂ§öÁöÑÂâç10ÁªÑ‰ΩúËÄÖ
top_edges = sorted(G.edges(data=True), key=lambda x: x[2]['weight'], reverse=True)[:50]
print("\nÂêà‰ΩúÊúÄÂ§öÁöÑÂâç50ÁªÑ‰ΩúËÄÖÔºö")
for u, v, d in top_edges:
    print(f"{u} - {v}Ôºö{d['weight']} Ê¨°")

# ÂèØËßÜÂåñÁΩëÁªúÔºàÈÄâÂâç100ËäÇÁÇπÊûÑÊàêÂ≠êÂõæÔºâ
top_nodes = sorted(G.degree, key=lambda x: x[1], reverse=True)[:100]
subG = G.subgraph([n for n, _ in top_nodes])

plt.figure(figsize=(12, 12))
nx.draw_networkx(subG, with_labels=True, node_size=80, font_size=6)
plt.title("‰ΩúËÄÖÂêà‰ΩúÁΩëÁªúÔºàTop 100 ‰ΩúËÄÖÔºâ")
plt.axis('off')
plt.show()
```

```python
import pandas as pd
import networkx as nx
from itertools import combinations

# ËØªÂèñ‰ΩúËÄÖÊï∞ÊçÆ
df = pd.read_csv('2authors.csv', sep=';')
df = df.dropna(subset=['‰ΩúËÄÖ', 'ÈìæÊé•'])

# ÂàõÂª∫Êó†ÂêëÂõæ
G = nx.Graph()

# ÊûÑÂª∫‰ΩúËÄÖÂêà‰ΩúËæπ
for link, group in df.groupby('ÈìæÊé•'):
    authors = group['‰ΩúËÄÖ'].unique()
    if len(authors) >= 2:
        for a1, a2 in combinations(authors, 2):
            if G.has_edge(a1, a2):
                G[a1][a2]['weight'] += 1
            else:
                G.add_edge(a1, a2, weight=1)

# ÂØºÂá∫ËäÇÁÇπÊï∞ÊçÆ
nodes_data = pd.DataFrame({
    "Id": list(G.nodes()),
    "Label": list(G.nodes()),
    "Degree": [G.degree(n) for n in G.nodes()]
})
nodes_data.to_csv("gephi_nodes.csv", index=False)

# ÂØºÂá∫ËæπÊï∞ÊçÆ
edges_data = pd.DataFrame([
    {"Source": u, "Target": v, "Weight": d["weight"]}
    for u, v, d in G.edges(data=True)
])
edges_data.to_csv("gephi_edges.csv", index=False)

print("Â∑≤ÊàêÂäüÂØºÂá∫ Gephi Ê†ºÂºèÁöÑËäÇÁÇπ‰∏éËæπÊï∞ÊçÆÔºögephi_nodes.csv Âíå gephi_edges.csv")
```

```python
import pandas as pd
import networkx as nx
from itertools import combinations
import matplotlib.pyplot as plt

# ËØªÂèñÂ∏¶Êúâ‚ÄúÂ§ßÂ≠¶‚ÄùÂàóÁöÑÊï∞ÊçÆ
df = pd.read_csv('2authors.csv', sep=';')

# Ê∏ÖÁêÜÊï∞ÊçÆÔºöÂè™‰øùÁïôÊúâÂ§ßÂ≠¶ÂíåÊñáÁ´†ÈìæÊé•ÁöÑË°å
df = df.dropna(subset=['ÈìæÊé•', 'Â§ßÂ≠¶'])

# ÂàùÂßãÂåñÂõæ
G_uni = nx.Graph()

# ÊåâÊñáÁ´†ÈìæÊé•ÂàÜÁªÑ
for link, group in df.groupby('ÈìæÊé•'):
    universities = group['Â§ßÂ≠¶'].dropna().unique()
    if len(universities) >= 2:
        for pair in combinations(sorted(universities), 2):
            u1, u2 = pair
            if G_uni.has_edge(u1, u2):
                G_uni[u1][u2]['weight'] += 1
            else:
                G_uni.add_edge(u1, u2, weight=1)

# ÊâìÂç∞ÁΩëÁªú‰ø°ÊÅØ
print(f"Â§ßÂ≠¶ËäÇÁÇπÊï∞ÈáèÔºö{G_uni.number_of_nodes()}")
print(f"Âêà‰ΩúËæπÊï∞ÈáèÔºö{G_uni.number_of_edges()}")

# ÊòæÁ§∫Âêà‰ΩúÊúÄÂ§öÁöÑÂâç20ÂØπÂ§ßÂ≠¶
top_uni_edges = sorted(G_uni.edges(data=True), key=lambda x: x[2]['weight'], reverse=True)[:20]
print("\nÂêà‰ΩúÊúÄÂ§öÁöÑÂâç20ÁªÑÂ§ßÂ≠¶Ôºö")
for u, v, d in top_uni_edges:
    print(f"{u} - {v}Ôºö{d['weight']} Ê¨°")

# ÂèØËßÜÂåñÂâç 50 ÊâÄÂ§ßÂ≠¶Âêà‰ΩúÁΩëÁªúÔºàÁÆÄÁï•ÁâàÔºâ
top_unis = sorted(G_uni.degree, key=lambda x: x[1], reverse=True)[:50]
subG = G_uni.subgraph([n for n, _ in top_unis])

plt.figure(figsize=(12, 12))
nx.draw_networkx(subG, with_labels=True, node_size=100, font_size=6)
plt.title("Â§ßÂ≠¶Âêà‰ΩúÁΩëÁªúÔºàTop 50Ôºâ")
plt.axis('off')
plt.show()
```

```python
import pandas as pd

# ËØªÂèñÊï∞ÊçÆ
df = pd.read_csv("5reviewers_rounds.csv", sep=";")

# ËΩ¨Êç¢‰∏∫ datetime
df["VersionÊó∂Èó¥"] = pd.to_datetime(df["VersionÊó∂Èó¥"], errors="coerce")

# ÂéªÈô§ÈáçÂ§çÁöÑË°åÔºàÁ°Æ‰øù‰∏ÄÁØáÊñáÁ´†‰∏Ä‰∏™ÁâàÊú¨‰∏ÄÊù°ËÆ∞ÂΩïÔºâ
df_unique = df.drop_duplicates(subset=["È¢òÁõÆ", "VersionÁºñÂè∑"])

# Âè™‰øùÁïôÂ§öËΩÆËØÑÂÆ°ÁöÑÊñáÁ´†
version_counts = df_unique.groupby("È¢òÁõÆ")["VersionÁºñÂè∑"].nunique().reset_index()
version_counts.columns = ["È¢òÁõÆ", "ËØÑÂÆ°ËΩÆÊ¨°Êï∞"]
multi_round_titles = version_counts[version_counts["ËØÑÂÆ°ËΩÆÊ¨°Êï∞"] > 1]["È¢òÁõÆ"]
df_multi = df_unique[df_unique["È¢òÁõÆ"].isin(multi_round_titles)].copy()

# ÊéíÂ∫èÂêéËÆ°ÁÆóÊó∂Èó¥Â∑Æ
df_multi = df_multi.sort_values(by=["È¢òÁõÆ", "VersionÁºñÂè∑"])
df_multi["ÁâàÊú¨Êó∂Èó¥Èó¥ÈöîÔºàÂ§©Ôºâ"] = df_multi.groupby("È¢òÁõÆ")["VersionÊó∂Èó¥"].diff().dt.days

# ÂéªÈô§Á¨¨‰∏ÄËΩÆÊó†Êó∂Èó¥Â∑ÆÁöÑÁ©∫ÂÄº
df_gap = df_multi.dropna(subset=["ÁâàÊú¨Êó∂Èó¥Èó¥ÈöîÔºàÂ§©Ôºâ"])

# ÂàÜÊûêÂπ≥ÂùáÈó¥Èöî
avg_gaps = df_gap.groupby("È¢òÁõÆ")["ÁâàÊú¨Êó∂Èó¥Èó¥ÈöîÔºàÂ§©Ôºâ"].mean()

# ÊúÄÂø´ vs ÊúÄÊÖ¢Ââç10
fastest_articles = avg_gaps.sort_values().head(10)
slowest_articles = avg_gaps.sort_values(ascending=False).head(10)

print("\n ÂÆ°ÂæóÊúÄÂø´ÁöÑÊñáÁ´† Top 10ÔºàÂπ≥ÂùáÈó¥ÈöîÂ§©Êï∞ÔºâÔºö")
print(fastest_articles)

print("\n ÂÆ°ÂæóÊúÄ‰πÖÁöÑÊñáÁ´† Top 10ÔºàÂπ≥ÂùáÈó¥ÈöîÂ§©Êï∞ÔºâÔºö")
print(slowest_articles)


# Êù•Ëá™Âì™‰∫õÂ≠¶ÁßëÂë¢
# ËØªÂèñÂ≠¶Áßë‰ø°ÊÅØ
df_meta = pd.read_csv("output.csv")
df_meta["È¢òÁõÆ"] = df_meta["È¢òÁõÆ"].astype(str).str.strip()

# ÂáÜÂ§áÔºöÊèêÂèñÂâç20ÂêçÈ¢òÁõÆ
fast_titles = avg_gaps.sort_values().head(10).index.tolist()
slow_titles = avg_gaps.sort_values(ascending=False).head(10).index.tolist()

# ÂêàÂπ∂Â≠¶Áßë‰ø°ÊÅØ
df_fast = df_meta[df_meta["È¢òÁõÆ"].isin(fast_titles)]
df_slow = df_meta[df_meta["È¢òÁõÆ"].isin(slow_titles)]

# Â≠¶ÁßëÁªüËÆ°
fast_discipline_counts = df_fast["Â≠¶Áßë"].value_counts()
slow_discipline_counts = df_slow["Â≠¶Áßë"].value_counts()

print("\nüìö ÂÆ°Á®øÊúÄÂø´ÁöÑÂâç10ÁØáÊñáÁ´†ÊâÄÂØπÂ∫îÁöÑÂ≠¶ÁßëÂàÜÂ∏ÉÔºö")
print(fast_discipline_counts)

print("\nüìö ÂÆ°Á®øÊúÄÊÖ¢ÁöÑÂâç10ÁØáÊñáÁ´†ÊâÄÂØπÂ∫îÁöÑÂ≠¶ÁßëÂàÜÂ∏ÉÔºö")
print(slow_discipline_counts)
```

```python
import pandas as pd

#  ËØªÂèñÊï∞ÊçÆÂπ∂Â§ÑÁêÜÊó∂Èó¥
df = pd.read_csv("5reviewers_rounds.csv", sep=";")
df["VersionÊó∂Èó¥"] = pd.to_datetime(df["VersionÊó∂Èó¥"], errors="coerce")

#  ÂéªÈáçÔºàÁ°Æ‰øù‰∏ÄÁØáÊñáÁ´†‰∏Ä‰∏™ÁâàÊú¨ÂîØ‰∏ÄÔºâ
df_unique = df.drop_duplicates(subset=["È¢òÁõÆ", "VersionÁºñÂè∑"])


# Â§ÑÁêÜÁâàÊú¨ÁºñÂè∑‰∏∫Êï∞Â≠óÔºàÂ¶Ç 'Version 1' -> 1Ôºâ
df_sorted = df_unique.sort_values(by=["È¢òÁõÆ", "VersionÊó∂Èó¥"])

# ÂÆâÂÖ®ÊèêÂèñÁâàÊú¨ÁºñÂè∑ÁöÑÊï∞Â≠ó
df_sorted["VersionÁºñÂè∑Êï∞Â≠ó"] = (
    df_sorted["VersionÁºñÂè∑"]
    .astype(str)
    .str.extract(r"(\d+)")
    .astype(float)  # Áî® float ‰øùÁïô NaN
)

# ËÆ°ÁÆóÁõ∏ÈÇªÁâàÊú¨Êó∂Èó¥Èó¥Èöî
df_sorted["Êó∂Èó¥Èó¥Èöî"] = df_sorted.groupby("È¢òÁõÆ")["VersionÊó∂Èó¥"].diff().dt.days
df_sorted["Ââç‰∏ÄËΩÆÁºñÂè∑"] = df_sorted.groupby("È¢òÁõÆ")["VersionÁºñÂè∑Êï∞Â≠ó"].shift()

# ‰ªÖ‰øùÁïôÊúâÂâç‰∏ÄËΩÆÁâàÊú¨ÂíåÁºñÂè∑Êï∞ÊçÆÁöÑËÆ∞ÂΩï
df_gaps = df_sorted.dropna(subset=["Êó∂Èó¥Èó¥Èöî", "Ââç‰∏ÄËΩÆÁºñÂè∑", "VersionÁºñÂè∑Êï∞Â≠ó"]).copy()

# ÊûÑÈÄ†ËΩÆÊ¨°Èó¥ÈöîÊ†áÁ≠æÔºöÂ¶Ç "1‚Üí2"
df_gaps["ËΩÆÊ¨°Èó¥Èöî"] = df_gaps["Ââç‰∏ÄËΩÆÁºñÂè∑"].astype(int).astype(str) + "‚Üí" + df_gaps["VersionÁºñÂè∑Êï∞Â≠ó"].astype(int).astype(str)


# ÂêÑËΩÆÊ¨°Èó¥ÁöÑÂπ≥ÂùáÁ≠âÂæÖÂ§©Êï∞
gap_summary = df_gaps.groupby("ËΩÆÊ¨°Èó¥Èöî")["Êó∂Èó¥Èó¥Èöî"].mean().sort_index()
print("\nüìä ÂêÑËΩÆÊ¨°‰πãÈó¥ÁöÑÂπ≥ÂùáÁ≠âÂæÖÂ§©Êï∞Ôºö")
print(gap_summary)

#  ËØªÂèñÂ≠¶Áßë‰ø°ÊÅØ
df_meta = pd.read_csv("output.csv")
df_meta["È¢òÁõÆ"] = df_meta["È¢òÁõÆ"].astype(str).str.strip()

#  ÂêàÂπ∂Â≠¶Áßë‰ø°ÊÅØÂà∞Èó¥ÈöîË°®
df_gaps = df_gaps.merge(df_meta[["È¢òÁõÆ", "Â≠¶Áßë"]], on="È¢òÁõÆ", how="left")

# ÊåâËΩÆÊ¨°ÂíåÂ≠¶ÁßëÁªüËÆ°Âπ≥ÂùáÁ≠âÂæÖÂ§©Êï∞
discipline_gap = df_gaps.groupby(["ËΩÆÊ¨°Èó¥Èöî", "Â≠¶Áßë"])["Êó∂Èó¥Èó¥Èöî"].mean().unstack().T

print("\nüìö ÂêÑÂ≠¶ÁßëÂú®‰∏çÂêåËΩÆÊ¨°‰πãÈó¥ÁöÑÂπ≥ÂùáÁ≠âÂæÖÂ§©Êï∞Ôºö")
print(discipline_gap)

# ÁªüËÆ°ÊØè‰∏™Â≠¶Áßë-ËΩÆÊ¨°ÁöÑÊñáÁ´†Êï∞Èáè
discipline_counts = df_gaps.groupby(["ËΩÆÊ¨°Èó¥Èöî", "Â≠¶Áßë"])["È¢òÁõÆ"].count().unstack().T

print("\nüìö ÂêÑÂ≠¶ÁßëÂú®‰∏çÂêåËΩÆÊ¨°‰πãÈó¥ÁöÑÊñáÁ´†Êï∞ÈáèÔºö")
print(discipline_counts)
```

```python
import pandas as pd

# ================================
# 1Ô∏è‚É£ ËØªÂèñÊï∞ÊçÆÂπ∂È¢ÑÂ§ÑÁêÜ
# ================================
df = pd.read_csv("7funders_extracted.csv", sep=";")
df["ÊñáÁ´†ID"] = df["ÈìæÊé•"].astype(str).str.extract(r'/articles/([^/]+)')
df["È¢òÁõÆ"] = df["È¢òÁõÆ"].astype(str).str.strip()

# ================================
# 2Ô∏è‚É£ Ê†áÂáÜÂåñËµÑÂä©ËÄÖÂêçÁß∞
# ================================
def normalize_funder_name(name):
    name = str(name).strip().lower()
    if any(k in name for k in ["horizon 2020", "h2020", "horizon2020"]):
        return "Horizon 2020"
    elif any(k in name for k in ["horizon europe", "heu", "horizoneurope"]):
        return "Horizon Europe"
    else:
        return name.title()  # È¶ñÂ≠óÊØçÂ§ßÂÜôÔºå‰øùÊåÅÊï¥Ê¥Å

df["Ê†áÂáÜÂåñËµÑÂä©ËÄÖÂêçÁß∞"] = df["ËµÑÂä©ËÄÖÂêçÁß∞"].fillna("").apply(normalize_funder_name)

# ================================
# 3Ô∏è‚É£ Ê∑ªÂä† Horizon È°πÁõÆÊ†áÂøóÂàó
# ================================
# ËÅöÂêàÊØèÁØáÊñáÁ´†ÊâÄÊúâÊ†áÂáÜÂåñÂêéÁöÑËµÑÂä©ËÄÖ
funders_per_article = df.groupby("ÊñáÁ´†ID")["Ê†áÂáÜÂåñËµÑÂä©ËÄÖÂêçÁß∞"].apply(lambda x: set(x.dropna()))

def classify_funding(funders):
    has_h2020 = "Horizon 2020" in funders
    has_heu = "Horizon Europe" in funders
    return pd.Series({
        "ÊòØÂê¶_H2020": has_h2020,
        "ÊòØÂê¶_HEurope": has_heu,
        "ÊòØÂê¶_ÂèåÈáçËµÑÂä©": has_h2020 and has_heu,
        "ÊòØÂê¶_ÂùáÊó†ËµÑÂä©": not has_h2020 and not has_heu
    })

horizon_flags = funders_per_article.apply(classify_funding)
df = df.merge(horizon_flags, left_on="ÊñáÁ´†ID", right_index=True, how="left")

# ================================
# 4Ô∏è‚É£ Âü∫Êú¨ÁªüËÆ°
# ================================
print(f"üìÑ ÊÄªËÆ∞ÂΩïÊï∞Ôºö{len(df)}")
print(f"üìÑ Áã¨ÁâπÊñáÁ´†Êï∞ÔºàÊåâÈ¢òÁõÆÔºâÔºö{df['È¢òÁõÆ'].nunique()}")

# ÊØèÁØáÊñáÁ´†Âè™‰øùÁïô‰∏ÄÊù°ËÆ∞ÂΩïÔºà‰ΩøÁî®ÊúÄÂ§ßËµÑÂä©ËÄÖ‰∏™Êï∞Ôºâ
funder_per_article = df.groupby("ÊñáÁ´†ID")["ËµÑÂä©ËÄÖ‰∏™Êï∞"].max()

# ÁÑ∂ÂêéÁªüËÆ°ÊØèÁßçËµÑÂä©ËÄÖ‰∏™Êï∞ÂØπÂ∫îÁöÑÁã¨ÁâπÊñáÁ´†Êï∞Èáè
funder_counts_unique = funder_per_article.value_counts().sort_index()

print("\nüìä ‰∏çÂêåËµÑÂä©ËÄÖÊï∞ÈáèÂØπÂ∫îÁöÑÔºàÁã¨ÁâπÔºâÊñáÁ´†Êï∞Ôºö")
print(funder_counts_unique)

# ================================
# 5Ô∏è‚É£ Top 20 ËµÑÂä©Êú∫ÊûÑÔºàÊ†áÂáÜÂåñÔºâ
# ================================
top_funders = df["Ê†áÂáÜÂåñËµÑÂä©ËÄÖÂêçÁß∞"].value_counts().head(20)
print("\nüèÜ Âá∫Áé∞Ê¨°Êï∞ÊúÄÂ§öÁöÑÊ†áÂáÜÂåñËµÑÂä©Êú∫ÊûÑ Top 20Ôºö")
print(top_funders)

# ================================
# 6Ô∏è‚É£ Êó†ËµÑÂä©ÁöÑÊñáÁ´†ÂàóË°®
# ================================
no_funding_df = df[df["ËµÑÂä©ËÄÖ‰∏™Êï∞"] == 0][["È¢òÁõÆ"]].drop_duplicates()
print("\nüìÑ Ê≤°Êúâ‰ªª‰ΩïËµÑÂä©ÁöÑÊñáÁ´†ÔºàÂâç10Ë°åÔºâÔºö")
print(no_funding_df.head(10))

# ================================
# 7Ô∏è‚É£ ÂêåÊó∂Êé•Âèó Horizon 2020 & Europe ÁöÑÊñáÁ´†
# ================================
both_df = df[df["ÊòØÂê¶_ÂèåÈáçËµÑÂä©"]][["ÊñáÁ´†ID", "È¢òÁõÆ", "Ê†áÂáÜÂåñËµÑÂä©ËÄÖÂêçÁß∞"]].drop_duplicates()
print(f"\nüåê ÂêåÊó∂Êé•Âèó Horizon 2020 Âíå Europe ËµÑÂä©ÁöÑÊñáÁ´†Êï∞Ôºö{both_df['ÊñáÁ´†ID'].nunique()}")
print(both_df.head(10))

# ================================
# 8Ô∏è‚É£ Êú™Êé•Âèó Horizon È°πÁõÆÁöÑÊñáÁ´† + Â≠¶Áßë
# ================================
# Âä†ËΩΩÂ≠¶Áßë‰ø°ÊÅØ
df_meta = pd.read_csv("output.csv")
df_meta["È¢òÁõÆ"] = df_meta["È¢òÁõÆ"].astype(str).str.strip()

# Á≠õÈÄâÊú™Êé•Âèó Horizon ‰ªª‰ΩïËµÑÂä©ÁöÑÊñáÁ´†
neither_df = df[df["ÊòØÂê¶_ÂùáÊó†ËµÑÂä©"]].copy()

# ËÅöÂêàÊØèÁØáÊñáÁ´†‰ø°ÊÅØ
grouped = neither_df.groupby("ÊñáÁ´†ID").agg({
    "È¢òÁõÆ": "first",
    "ÈìæÊé•": "first",
    "Ê†áÂáÜÂåñËµÑÂä©ËÄÖÂêçÁß∞": lambda x: "; ".join(sorted(set(x.dropna()))),
    "ËµÑÂä©ËÄÖ‰∏™Êï∞": "max"
}).reset_index()

# ÂêàÂπ∂Â≠¶Áßë‰ø°ÊÅØ
final_df = pd.merge(grouped, df_meta[["È¢òÁõÆ", "Â≠¶Áßë"]], on="È¢òÁõÆ", how="left")

# Â±ïÁ§∫ÂâçÂá†Ë°å
print("\nüìö Êú™Êé•Âèó Horizon È°πÁõÆËµÑÂä©ÁöÑÊñáÁ´†ÔºàÂê´Â≠¶Áßë‰ø°ÊÅØÔºâÂâç10Ë°åÔºö")
print(final_df.head(10))

# ================================
# 9Ô∏è‚É£ ÂèØÈÄâÔºöÂØºÂá∫Ê∏ÖÊ¥óÂêéÁöÑÊï∞ÊçÆ
# ================================
df.to_csv("7funders_cleaned.csv", index=False)
final_df.to_csv("articles_without_horizon.csv", index=False)
```

```python
import pandas as pd
import re

# 1. Ê†áÂáÜÂåñÂáΩÊï∞
def standardize_title(title):
    if pd.isna(title):
        return ''
    title = title.lower()
    title = re.sub(r'\s+', '', title)
    title = re.sub(r'[^\w]', '', title)
    return title

# 2. Âä†ËΩΩ Horizon ‰∏§‰∏™Êñá‰ª∂
df_h2020 = pd.read_csv('2020projectPublications.csv', sep=';', engine='python', on_bad_lines='skip')
df_heu = pd.read_csv('HORIZONEUROPEprojectPublications.csv', sep=';', engine='python', on_bad_lines='skip')
df_all = pd.concat([df_h2020, df_heu], ignore_index=True)
df_all = df_all[df_all['title'].notna()]  # ÂéªÈô§Êó†Ê†áÈ¢òËÆ∞ÂΩï
df_all['title_std'] = df_all['title'].apply(standardize_title)

# 3. Âä†ËΩΩ ORE Âπ≥Âè∞ output.csvÔºà‰Ω†Áà¨‰∏ãÊù•ÁöÑÔºâ
df_output = pd.read_csv('output.csv')
df_output = df_output[df_output['È¢òÁõÆ'].notna()]
df_output['title_std'] = df_output['È¢òÁõÆ'].apply(standardize_title)

# 4. Áî®ÈõÜÂêà‰∫§ÈõÜËøõË°åÂåπÈÖç
ore_titles_std_set = set(df_output['title_std'].unique())
horizon_titles_std_set = set(df_all['title_std'].unique())

intersection = ore_titles_std_set & horizon_titles_std_set

print(f"OREÂπ≥Âè∞ÊñáÁ´†ÊÄªÊï∞Ôºö{len(ore_titles_std_set)}")
print(f"Âá∫Áé∞Âú® Horizon ÁôªËÆ∞Á≥ªÁªü‰∏≠ÁöÑ ORE ÊñáÁ´†Êï∞Ôºö{len(intersection)}")

# 5. ËÆ°ÁÆóË¶ÜÁõñÁéá
coverage = len(intersection) / len(ore_titles_std_set) * 100
print(f"Ë¢´ Horizon Êï∞ÊçÆË¶ÜÁõñÁöÑÊØî‰æãÔºö{coverage:.2f}%")
```
