```python
import pandas as pd
import pycountry
import pycountry_convert as pc

# 加载你上传的文件
df = pd.read_excel("作者国家分布.xlsx")

# 国家名称纠错表
country_corrections = {
    "UK": "United Kingdom",
    "USA": "United States",
    "Russia": "Russian Federation",
    "South Korea": "Korea, Republic of",
    "North Korea": "Korea, Democratic People's Republic of",
    "Iran": "Iran, Islamic Republic of",
    "Venezuela": "Venezuela, Bolivarian Republic of",
    "Syria": "Syrian Arab Republic",
    "Vietnam": "Viet Nam",
    "Bolivia": "Bolivia, Plurinational State of",
    "Tanzania": "Tanzania, United Republic of"
}

def get_continent(country_name):
    try:
        name = country_corrections.get(country_name, country_name)
        country = pycountry.countries.get(name=name)
        if not country:
            country = pycountry.countries.search_fuzzy(name)[0]
        code = country.alpha_2
        continent_code = pc.country_alpha2_to_continent_code(code)
        return pc.convert_continent_code_to_continent_name(continent_code)
    except:
        return "Unknown"

# 添加大洲列
df["Continent"] = df["Country"].apply(get_continent)

# 保存结果
df.to_excel("国家_大洲_统计结果.xlsx", index=False)
```

```python
import pandas as pd
import networkx as nx
from itertools import combinations
import matplotlib.pyplot as plt

# 读取数据
df = pd.read_csv('2authors.csv', sep=';')  # 根据你的文件格式调整分隔符

# 去除缺失项
df = df.dropna(subset=['作者', '链接'])

# 初始化图
G = nx.Graph()

# 按文章链接分组，构建合作边
for link, group in df.groupby('链接'):
    authors = group['作者'].unique()
    if len(authors) >= 2:
        for pair in combinations(authors, 2):
            a1, a2 = sorted(pair)
            if G.has_edge(a1, a2):
                G[a1][a2]['weight'] += 1
            else:
                G.add_edge(a1, a2, weight=1)

# 打印合作网络信息
print(f"作者节点数量：{G.number_of_nodes()}")
print(f"合作边数量：{G.number_of_edges()}")

# 显示合作最多的前10组作者
top_edges = sorted(G.edges(data=True), key=lambda x: x[2]['weight'], reverse=True)[:50]
print("\n合作最多的前50组作者：")
for u, v, d in top_edges:
    print(f"{u} - {v}：{d['weight']} 次")

# 可视化网络（选前100节点构成子图）
top_nodes = sorted(G.degree, key=lambda x: x[1], reverse=True)[:100]
subG = G.subgraph([n for n, _ in top_nodes])

plt.figure(figsize=(12, 12))
nx.draw_networkx(subG, with_labels=True, node_size=80, font_size=6)
plt.title("作者合作网络（Top 100 作者）")
plt.axis('off')
plt.show()
```

```python
import pandas as pd
import networkx as nx
from itertools import combinations

# 加载数据
df = pd.read_csv('2authors.csv', sep=';')
df = df.dropna(subset=['作者', '链接'])

# 构建原始合作图
G = nx.Graph()
for link, group in df.groupby('链接'):
    authors = group['作者'].unique()
    if len(authors) >= 2:
        for a1, a2 in combinations(authors, 2):
            if G.has_edge(a1, a2):
                G[a1][a2]['weight'] += 1
            else:
                G.add_edge(a1, a2, weight=1)

#  提取最大连通子图
largest_cc = max(nx.connected_components(G), key=len)
G_sub = G.subgraph(largest_cc).copy()

# 节点导出
nodes_df = pd.DataFrame({
    "Id": list(G_sub.nodes()),
    "Label": list(G_sub.nodes()),
    "Degree": [G_sub.degree(n) for n in G_sub.nodes()]
})
nodes_df.to_csv("gephi_nodes_largest_component.csv", index=False)

# 边导出
edges_df = pd.DataFrame([
    {"Source": u, "Target": v, "Weight": d["weight"]}
    for u, v, d in G_sub.edges(data=True)
])
edges_df.to_csv("gephi_edges_largest_component.csv", index=False)

print(f" 已成功导出最大连通子图：节点数 {G_sub.number_of_nodes()}，边数 {G_sub.number_of_edges()}")
```

```python
import pandas as pd
import networkx as nx
from itertools import combinations

# 加载数据
df = pd.read_csv("2authors.csv", sep=";")
df = df.dropna(subset=["作者", "链接", "标题"])

# === 构建完整合作图 ===
G = nx.Graph()
for link, group in df.groupby("链接"):
    authors = group["作者"].unique()
    for a1, a2 in combinations(authors, 2):
        if G.has_edge(a1, a2):
            G[a1][a2]["weight"] += 1
        else:
            G.add_edge(a1, a2, weight=1, link=link)

# === 找到 Pamela Klaassen 所在的连通子图 ===
target_author = "Pamela Klaassen"
if target_author in G:
    component_nodes = nx.node_connected_component(G, target_author)
    subG = G.subgraph(component_nodes).copy()
else:
    print("目标作者不在合作网络中")
    subG = nx.Graph()

# === 导出 Gephi 文件 ===
nodes_df = pd.DataFrame({
    "Id": list(subG.nodes()),
    "Label": list(subG.nodes()),
    "Degree": [subG.degree(n) for n in subG.nodes()]
})
edges_df = pd.DataFrame([
    {"Source": u, "Target": v, "Weight": d["weight"]}
    for u, v, d in subG.edges(data=True)
])
nodes_df.to_csv("Pamela_network_nodes.csv", index=False)
edges_df.to_csv("Pamela_network_edges.csv", index=False)

# === 找出这些人参与的全部文章（标题 + 作者） ===
people_in_subgraph = set(subG.nodes())
df_sub_articles = df[df["作者"].isin(people_in_subgraph)]
article_info = df_sub_articles[["链接", "标题", "作者"]].drop_duplicates()
article_info.to_excel("Pamela_network_articles.xlsx", index=False)

print(f" 共导出网络节点 {len(nodes_df)} 个，边 {len(edges_df)} 条，文章数约 {article_info['链接'].nunique()} 篇")
```

```python
import pandas as pd
import networkx as nx
from itertools import combinations
import matplotlib.pyplot as plt

# 读取带有“大学”列的数据
df = pd.read_csv('2authors.csv', sep=';')

# 清理数据：只保留有大学和文章链接的行
df = df.dropna(subset=['链接', '大学'])

# 初始化图
G_uni = nx.Graph()

# 按文章链接分组
for link, group in df.groupby('链接'):
    universities = group['大学'].dropna().unique()
    if len(universities) >= 2:
        for pair in combinations(sorted(universities), 2):
            u1, u2 = pair
            if G_uni.has_edge(u1, u2):
                G_uni[u1][u2]['weight'] += 1
            else:
                G_uni.add_edge(u1, u2, weight=1)

# 打印网络信息
print(f"大学节点数量：{G_uni.number_of_nodes()}")
print(f"合作边数量：{G_uni.number_of_edges()}")

# 显示合作最多的前20对大学
top_uni_edges = sorted(G_uni.edges(data=True), key=lambda x: x[2]['weight'], reverse=True)[:20]
print("\n合作最多的前20组大学：")
for u, v, d in top_uni_edges:
    print(f"{u} - {v}：{d['weight']} 次")

# 可视化前 50 所大学合作网络（简略版）
top_unis = sorted(G_uni.degree, key=lambda x: x[1], reverse=True)[:50]
subG = G_uni.subgraph([n for n, _ in top_unis])

plt.figure(figsize=(12, 12))
nx.draw_networkx(subG, with_labels=True, node_size=100, font_size=6)
plt.title("大学合作网络（Top 50）")
plt.axis('off')
plt.show()
```

```python
import pandas as pd

# 读取数据
df = pd.read_csv("5reviewers_rounds.csv", sep=";")

# 转换为 datetime
df["Version时间"] = pd.to_datetime(df["Version时间"], errors="coerce")

# 去除重复的行（确保一篇文章一个版本一条记录）
df_unique = df.drop_duplicates(subset=["题目", "Version编号"])

# 只保留多轮评审的文章
version_counts = df_unique.groupby("题目")["Version编号"].nunique().reset_index()
version_counts.columns = ["题目", "评审轮次数"]
multi_round_titles = version_counts[version_counts["评审轮次数"] > 1]["题目"]
df_multi = df_unique[df_unique["题目"].isin(multi_round_titles)].copy()

# 排序后计算时间差
df_multi = df_multi.sort_values(by=["题目", "Version编号"])
df_multi["版本时间间隔（天）"] = df_multi.groupby("题目")["Version时间"].diff().dt.days

# 去除第一轮无时间差的空值
df_gap = df_multi.dropna(subset=["版本时间间隔（天）"])

# 分析平均间隔
avg_gaps = df_gap.groupby("题目")["版本时间间隔（天）"].mean()

# 最快 vs 最慢前10
fastest_articles = avg_gaps.sort_values().head(10)
slowest_articles = avg_gaps.sort_values(ascending=False).head(10)

print("\n 审得最快的文章 Top 10（平均间隔天数）：")
print(fastest_articles)

print("\n 审得最久的文章 Top 10（平均间隔天数）：")
print(slowest_articles)


# 来自哪些学科呢
# 读取学科信息
df_meta = pd.read_csv("output.csv")
df_meta["题目"] = df_meta["题目"].astype(str).str.strip()

# 准备：提取前20名题目
fast_titles = avg_gaps.sort_values().head(10).index.tolist()
slow_titles = avg_gaps.sort_values(ascending=False).head(10).index.tolist()

# 合并学科信息
df_fast = df_meta[df_meta["题目"].isin(fast_titles)]
df_slow = df_meta[df_meta["题目"].isin(slow_titles)]

# 学科统计
fast_discipline_counts = df_fast["学科"].value_counts()
slow_discipline_counts = df_slow["学科"].value_counts()

print("\n📚 审稿最快的前10篇文章所对应的学科分布：")
print(fast_discipline_counts)

print("\n📚 审稿最慢的前10篇文章所对应的学科分布：")
print(slow_discipline_counts)
```

```python
import pandas as pd

#  读取数据并处理时间
df = pd.read_csv("5reviewers_rounds.csv", sep=";")
df["Version时间"] = pd.to_datetime(df["Version时间"], errors="coerce")

#  去重（确保一篇文章一个版本唯一）
df_unique = df.drop_duplicates(subset=["题目", "Version编号"])


# 处理版本编号为数字（如 'Version 1' -> 1）
df_sorted = df_unique.sort_values(by=["题目", "Version时间"])

# 安全提取版本编号的数字
df_sorted["Version编号数字"] = (
    df_sorted["Version编号"]
    .astype(str)
    .str.extract(r"(\d+)")
    .astype(float)  # 用 float 保留 NaN
)

# 计算相邻版本时间间隔
df_sorted["时间间隔"] = df_sorted.groupby("题目")["Version时间"].diff().dt.days
df_sorted["前一轮编号"] = df_sorted.groupby("题目")["Version编号数字"].shift()

# 仅保留有前一轮版本和编号数据的记录
df_gaps = df_sorted.dropna(subset=["时间间隔", "前一轮编号", "Version编号数字"]).copy()

# 构造轮次间隔标签：如 "1→2"
df_gaps["轮次间隔"] = df_gaps["前一轮编号"].astype(int).astype(str) + "→" + df_gaps["Version编号数字"].astype(int).astype(str)


# 各轮次间的平均等待天数
gap_summary = df_gaps.groupby("轮次间隔")["时间间隔"].mean().sort_index()
print("\n📊 各轮次之间的平均等待天数：")
print(gap_summary)

#  读取学科信息
df_meta = pd.read_csv("output.csv")
df_meta["题目"] = df_meta["题目"].astype(str).str.strip()

#  合并学科信息到间隔表
df_gaps = df_gaps.merge(df_meta[["题目", "学科"]], on="题目", how="left")

# 按轮次和学科统计平均等待天数
discipline_gap = df_gaps.groupby(["轮次间隔", "学科"])["时间间隔"].mean().unstack().T

print("\n📚 各学科在不同轮次之间的平均等待天数：")
print(discipline_gap)

# 统计每个学科-轮次的文章数量
discipline_counts = df_gaps.groupby(["轮次间隔", "学科"])["题目"].count().unstack().T

print("\n📚 各学科在不同轮次之间的文章数量：")
print(discipline_counts)
```

```python
import pandas as pd

# ================================
# 1️⃣ 读取数据并预处理
# ================================
df = pd.read_csv("7funders_extracted.csv", sep=";")
df["文章ID"] = df["链接"].astype(str).str.extract(r'/articles/([^/]+)')
df["题目"] = df["题目"].astype(str).str.strip()

# ================================
# 2️⃣ 标准化资助者名称
# ================================
def normalize_funder_name(name):
    name = str(name).strip().lower()
    if any(k in name for k in ["horizon 2020", "h2020", "horizon2020"]):
        return "Horizon 2020"
    elif any(k in name for k in ["horizon europe", "heu", "horizoneurope"]):
        return "Horizon Europe"
    else:
        return name.title()  # 首字母大写，保持整洁

df["标准化资助者名称"] = df["资助者名称"].fillna("").apply(normalize_funder_name)

# ================================
# 3️⃣ 添加 Horizon 项目标志列
# ================================
# 聚合每篇文章所有标准化后的资助者
funders_per_article = df.groupby("文章ID")["标准化资助者名称"].apply(lambda x: set(x.dropna()))

def classify_funding(funders):
    has_h2020 = "Horizon 2020" in funders
    has_heu = "Horizon Europe" in funders
    return pd.Series({
        "是否_H2020": has_h2020,
        "是否_HEurope": has_heu,
        "是否_双重资助": has_h2020 and has_heu,
        "是否_均无资助": not has_h2020 and not has_heu
    })

horizon_flags = funders_per_article.apply(classify_funding)
df = df.merge(horizon_flags, left_on="文章ID", right_index=True, how="left")

# ================================
# 4️⃣ 基本统计
# ================================
print(f"📄 总记录数：{len(df)}")
print(f"📄 独特文章数（按题目）：{df['题目'].nunique()}")

# 每篇文章只保留一条记录（使用最大资助者个数）
funder_per_article = df.groupby("文章ID")["资助者个数"].max()

# 然后统计每种资助者个数对应的独特文章数量
funder_counts_unique = funder_per_article.value_counts().sort_index()

print("\n📊 不同资助者数量对应的（独特）文章数：")
print(funder_counts_unique)

# ================================
# 5️⃣ Top 20 资助机构（标准化）
# ================================
top_funders = df["标准化资助者名称"].value_counts().head(20)
print("\n🏆 出现次数最多的标准化资助机构 Top 20：")
print(top_funders)

# ================================
# 6️⃣ 无资助的文章列表
# ================================
no_funding_df = df[df["资助者个数"] == 0][["题目"]].drop_duplicates()
print("\n📄 没有任何资助的文章（前10行）：")
print(no_funding_df.head(10))

# ================================
# 7️⃣ 同时接受 Horizon 2020 & Europe 的文章
# ================================
both_df = df[df["是否_双重资助"]][["文章ID", "题目", "标准化资助者名称"]].drop_duplicates()
print(f"\n🌐 同时接受 Horizon 2020 和 Europe 资助的文章数：{both_df['文章ID'].nunique()}")
print(both_df.head(10))

# ================================
# 8️⃣ 未接受 Horizon 项目的文章 + 学科
# ================================
# 加载学科信息
df_meta = pd.read_csv("output.csv")
df_meta["题目"] = df_meta["题目"].astype(str).str.strip()

# 筛选未接受 Horizon 任何资助的文章
neither_df = df[df["是否_均无资助"]].copy()

# 聚合每篇文章信息
grouped = neither_df.groupby("文章ID").agg({
    "题目": "first",
    "链接": "first",
    "标准化资助者名称": lambda x: "; ".join(sorted(set(x.dropna()))),
    "资助者个数": "max"
}).reset_index()

# 合并学科信息
final_df = pd.merge(grouped, df_meta[["题目", "学科"]], on="题目", how="left")

# 展示前几行
print("\n📚 未接受 Horizon 项目资助的文章（含学科信息）前10行：")
print(final_df.head(10))

# ================================
# 9️⃣ 可选：导出清洗后的数据
# ================================
df.to_csv("7funders_cleaned.csv", index=False)
final_df.to_csv("articles_without_horizon.csv", index=False)
```

```python
import pandas as pd
import re

# 1. 标准化函数
def standardize_title(title):
    if pd.isna(title):
        return ''
    title = title.lower()
    title = re.sub(r'\s+', '', title)
    title = re.sub(r'[^\w]', '', title)
    return title

# 2. 加载 Horizon 两个文件
df_h2020 = pd.read_csv('2020projectPublications.csv', sep=';', engine='python', on_bad_lines='skip')
df_heu = pd.read_csv('HORIZONEUROPEprojectPublications.csv', sep=';', engine='python', on_bad_lines='skip')
df_all = pd.concat([df_h2020, df_heu], ignore_index=True)
df_all = df_all[df_all['title'].notna()]  # 去除无标题记录
df_all['title_std'] = df_all['title'].apply(standardize_title)

# 3. 加载 ORE 平台 output.csv（你爬下来的）
df_output = pd.read_csv('output.csv')
df_output = df_output[df_output['题目'].notna()]
df_output['title_std'] = df_output['题目'].apply(standardize_title)

# 4. 用集合交集进行匹配
ore_titles_std_set = set(df_output['title_std'].unique())
horizon_titles_std_set = set(df_all['title_std'].unique())

intersection = ore_titles_std_set & horizon_titles_std_set

print(f"ORE平台文章总数：{len(ore_titles_std_set)}")
print(f"出现在 Horizon 登记系统中的 ORE 文章数：{len(intersection)}")

# 5. 计算覆盖率
coverage = len(intersection) / len(ore_titles_std_set) * 100
print(f"被 Horizon 数据覆盖的比例：{coverage:.2f}%")
```
